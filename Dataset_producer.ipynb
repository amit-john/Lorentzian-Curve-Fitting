```python
#loading data
file_name=input("Enter the filename")
data=pd.read_csv(str(file_name))
```

```python
#preprocessing of the dataset
l=["ppm"]
for i in range(1,data.shape[1]):
  l+=[str(i)]

data=pd.read_csv(str(file_name),names=l)
```

```python
#Checking the horizontal view of the plots
for i in range(1,data.shape[1]):
  plt.plot(data["ppm"],data[str(i)])
plt.show()
```

```python
#For doing one peak curve fitting
#lorentzian function for one peaK
def lorentzian_1(x,A1,x1,r1):
  L= (A1*r1)/(((x-x1)**2+r1**2)*np.pi)
  return L

#Normalising lorentzian function for one peak
def lorentzian(x,A1,x1,r1,slope):
  L= (A1*r1)/(((x-x1)**2+r1**2)*np.pi)/slope
  return L


#lorentzian function for obtaining the parameters for one peak
def Intensity_paramter(iter):
  #loading the data
  x=data["ppm"]
  y=data[str(iter)]

  #calculating the amplitude parmeter for the lorentzian function
  A=max(y)-((y[0]+y[len(y)-1])/2)

  #calculating the peak position
  for i in range(len(y)):
    if y[i]==max(y):
      x1=x[i]

  #calculating the fwhm
  hm=[]
  for _ in range(len(y)):
    hm+=[A/2]
  df=pd.DataFrame(hm)
  df

  Line_1=LineString(np.column_stack((x,y)))
  Line_2=LineString(np.column_stack((x,df)))
  intersection=Line_1.intersection(Line_2)
  x_co=[]
  y_co=[]
  #print(intersection)

  for i in intersection.geoms:
    x_co+=[i.x]
    y_co+=[i.y]
  r=abs(x_co[0]-x_co[1])
  #plt.axhline(y = A/2, color = 'r', linestyle = '-')
  #plt.plot(x,y)
  #plt.scatter(x=x_co,y=y_co)
  #plt.show()
  return A,r,x1,x,y



def fitting_function(iter):
  #obtaining the parameters
  A,r,x1,x,y=Intensity_paramter(iter)
  #obaiting the lorentzian
  y_fit_ini=lorentzian_1(x,A,x1,r/2)
  #normalising the lorentzian
  for i in range(len(y_fit_ini)):
    if y[i]==max(y):
      #print(y[i],y_fit_ini[i])
      slope=y_fit_ini[i]/y[i]
  #Obtaining the final fitted coordinates
  y_fit=lorentzian(x,A,x1,r/2,slope)
  #obtaing the plot
  fig = go.Figure()
  fig.add_trace(go.Scatter(x=x, y=y, mode='markers',name="data_points"))
  fig.add_trace(go.Scatter(x=x, y=y_fit, mode='lines',name="fitted_plot"))
  fig.show()
  return max(y_fit)
lis=[]
x_lis=[]
for i in range(1,data.shape[1]):
  try:
    print(i)
    y_coord=fitting_function(i)
    print(y_coord)
    lis+=[y_coord]
  except:
    #lis+=[max(data[str(i)])]
    lis+=[0]
    plt.plot(data["ppm"],data[str(i)])
    plt.show()
    print(max(data[str(i)]))
    print("Cannot be curve fitted")

for i in range(len(lis)):
  x_lis+=[i+1]
```

```python
#saving the fitted values (Max value only)
y_lis=[]
for i in range(1,data.shape[1]):
  y_lis+=[max(data[str(i)])]

print(y_lis)
print(lis)
```

```python
#loading the offset values
file=input("Enter the name of the file containg the offset frequencies: ")
f=open(file,"r")
frequency=float(input("Enter the frequency "))
freq=[]
for i in f:
  freq+=[int(i)/frequency]
print(freq,len(freq))
```

```python
#avg
avg=[]
for i in range(len(y_lis)-1,len(y_lis)-11,-1):
  avg+=[y_lis[i]]
ag=np.average(avg)
```

```python
#Implemeneting three pool curve fitting
filename = "final_data.csv"
# field names
#fields = ["offset","Intensities"]
rows = []
for i in range(2,len(y_lis)):
  rows+=[[freq[i],y_lis[i]/ag]]
# writing to csv file


with open(filename, 'w') as csvfile:
    # creating a csv writer object
    csvwriter = csv.writer(csvfile)
    # writing the data rows
    csvwriter.writerows(rows)
```
